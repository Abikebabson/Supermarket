#### To combine the files together as one Table

```python
supermarket = pd.concat([Lag, Abj, Ph])
```

```python
#checking the first few values in the table,
#to understand how to clean  the data
supermarket.head()
```

```python
#checking the last few values in the table,
#to understand how to clean  the data
supermarket.tail()
```

```python
#checking the columns in the table that needs,
#cleaning, few ones that needs to be renamed 
supermarket.columns
```

```python
supermarket.rename({'Invoice ID':'Invoice_Id', 'Customer type':'Customer_type', 'Product line':'Product_line', 'Unit price':'Unit_price','Tax 5%':'Tax_5%', 'gross margin percentage': 
                            'Gross_Mar_%', 'gross income': 'Gross_income'}, axis = 1, inplace = True)
```

```python
supermarket['Unit_price'] = supermarket['Unit_price'].str.replace(',', '')
```

```python
supermarket['Unit_price'] = supermarket['Unit_price'].astype(float)
```

```python
#checking if the columns have been renamed properly
supermarket.head()
```

```python
#checking if the table has a consistent number of,
#rows and columns of each table b4 it was concatenated
supermarket.shape
```

```python
#checking the data info to see which is in str instead of,
#float and integars, so cleaning needs to be done
supermarket.info()
```

```python
#checking for null values in the table,
supermarket.isna()
```

```python
#no null values in the data
supermarket.isna().sum()
```

```python
supermarket.describe()
```

```python
#changing date from str to real date time
supermarket['Day'] = supermarket['Date'].dt.day
supermarket['Month'] = supermarket['Date'].dt.month
supermarket['Year'] = supermarket['Date'].dt.year
supermarket['Hour'] = supermarket['Time'].dt.hour
```

### Visualization

```python
sns.countplot(x='Branch', data=supermarket).set_title('Branch Count')
```

```python
sns.countplot(x='City', data=supermarket).set_title('Branch Count')
```

### It is observed that there are more outlets in Branch A which is Lagos,followed by Branch B which is Portharcourt and Branch C being the least outlet which is Abuja.


```python
sns.countplot(y='Product_line', data = supermarket)
```

### This shows that the Product Lines are mostly purchased in descending order.

```python
sns.countplot(x= "Payment", hue = "Branch", data=supermarket).set_title("Payment Channel per Branch")
```

### This shows that customers means of payment at BRANCH A is the electronic form of payment(EPAY PAYMENT), while CARD PAYMENT CHANNEL is mostly use in BRANCH B and CASH PAYMENT is the highest method of payment in BRANCH C.

### Seeing the different method of payment used frequently in the different Branches. This form of payment should be made available at all times in those branches to enable the free flow of business


### Branch C which uses CASH PAYMENT mostly, it should be closely monitored to avoid theft and Audit Officers should be available to audit at all times to prevent missing funds, and to ensure money is kept safe and paid at banks regularly.

```python
sns.catplot(x='Product_line', y='Total', hue='Gender', data=supermarket, kind='bar', aspect=3)
```

### This shows that the Female bought more of the product lines of Home and lifestyle, Food and bevearages, and Fashion accessories, while the Male bought more of the product lines of Health and beauty, Sports and Travel.

### Although, both the Female and male had a very close range with Electronics accessories.

### With this observation, when a female enters their store they should know the product lines that should be advertise to them, that will catch their fancy, likewise the Male

```python
sns.boxplot(x='Branch', y="Rating", data=supermarket).set_title("Branch Rating")
```
